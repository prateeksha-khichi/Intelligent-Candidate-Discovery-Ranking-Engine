# Manual Validation Loop

Run this notebook to manually compare the formula ranking vs a human ranking for the 50 sample candidates.

```
import sys
import os
sys.path.append(os.path.abspath('..'))

import json
from src.ranking.scoring_engine import calculate_final_score
from src.offline.honeypot_detector import is_honeypot

def validate():
    with open('../data/raw/sample_candidates.json', 'r') as f:
        sample_candidates = json.load(f)
        
    print(f"Loaded {len(sample_candidates)} sample candidates.")
    
    scored = []
    for cand in sample_candidates:
        hp = is_honeypot(cand)
        # Mock semantic score for now since we don't have JD embedding loaded here
        sem_score = 0.8
        score = calculate_final_score(cand, sem_score, hp)
        scored.append((cand['candidate_id'], cand['profile'].get('current_title'), score))
        
    scored.sort(key=lambda x: x[2], reverse=True)
    
    for rank, item in enumerate(scored[:10], 1):
        print(f"{rank}. {item[0]} - {item[1]} (Score: {item[2]:.3f})")

validate()
```
